```python
def add(list_a, list_b):
    result = []
    for first, second in zip(list_a, list_b):
        result.append(first + second)
    x = 0
    for i in result:
        x += i
    return x

print(add(a,b))
```

```python
def getSum(myList):
  #variable sum
  sum = 0
  for item in myList:
    sum = sum+item
  return sum
getSum([1,2,3])

#sum = 0
#sum ditambah oleh item yang ada pada myList
# 0+1+2+3
# hasilnya 6
```

```python
def getSubstract(myList):
  #variable substract
  substract = 0
  for item in myList:
    substract = substract-item
  return substract
getSubstract([1,2,3])

#substract = 0
#substract dikurangi oleh item yang ada pada myList
# 0-1-2-3
# hasilnya -6
```

```python
def getDivide(myList):
  #variable divide
  divide = 1
  for item in myList:
    divide = divide/item
  return divide
getDivide([1,2,5])

#divide = 1
#divide dibagi oleh item yang ada pada myList
# 1/1/2/5
# hasilnya 0.1
```

```python
def getSumHimpunan(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum += item
  return sum
getSumHimpunan([[1,2,3],[10,20,30]])

#sum = 0
#sum ditambah oleh item setiap list
#list 1 berjumlah 6
#list 2 berjumlah 60
# 0+6+60
# 66
```

```python
def getDivideHimpunan(myList):
    results = []
    for row in myList:
        divide = row[0]
        for item in row[1:]:
            divide = divide / item
        results.append(divide)

    final = results[0]
    for val in results[1:]:
        final = final / val
    return final

getDivideHimpunan([[1,1,1],[1,2,5]])

#divide = row/list 1
#nilai divide akan dibagi oleh row selanjutnya / row 2
#List 1 berjumlah 1
#list 2 berjumlah 0.1
# 1/0.1
# 10
```

```python
def getSubstractHimpunan(myList):
    results = []
    for row in myList:
        sub = row[0]
        for item in row[1:]:
            sub -= item
        results.append(sub)

    result = results[0]
    for val in results[1:]:
        result -= val
    return result

getSubstractHimpunan([[1,2,3],[10,20,30]])

#list 1 berjumlah 1-2-3 = -4
#list 2 berjumlah 10-20-30 = -40
# list1-list2
# -4-(-40)
# 36
```

```python
def searchBinary(myList, item):
    first = 0
    last = len(myList) - 1
    foundFlag = False

    while first <= last and not foundFlag:
        mid = (first + last) // 2
        if myList[mid] == item:
            foundFlag = True
        else:
            if item < myList[mid]:
                last = mid - 1
            else:
                first = mid + 1
    return foundFlag
print(searchBinary([8, 9, 10, 100, 1000, 2000, 3000], 10))
print(searchBinary([8, 9, 10, 100, 1000, 2000, 3000], 5))
```
